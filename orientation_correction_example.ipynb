# Orientation Correction Example

This jupyter notebook demos how to use the ```orientation_correction.py``` module.

````python
import cv2
from modules.orientation_correction import OrientationCorrection
import os
from IPython.display import Image
````

Input parameter: Input image as numpy array (grayscale)

Output: Image with corrected orientation as numpy array

````python
# Input
input_image_path = 'assets/orientation.jpg'
img = cv2.imread(input_image_path, cv2.IMREAD_GRAYSCALE)
````

````python
# Show the image
Image(input_image_path)
````

````python
# This will correct the orientation of the image
corrected_orientation = OrientationCorrection(img).orientation_correction()
````

````python
# Save the image
cv2.imwrite('output.png', corrected_orientation)
````

## Output Image

````python
# Show the output
Image('Output.png')
````
